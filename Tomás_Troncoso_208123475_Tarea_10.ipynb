```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def met_euler(f, a, b, yo, N):
    """
    Aproxima la solución de una ecuación diferencial ordinaria (EDO) de primer orden utilizando el método de Euler.

    Parámetros:
    - f: Función invocable que representa la EDO, f(t, y), donde t es la variable independiente e y es la variable dependiente.
    - a: Valor inicial de la variable independiente.
    - b: Valor final de la variable independiente.
    - yo: Valor inicial de la variable dependiente en t=a.
    - N: Número de pasos para discretizar el intervalo [a, b].

    Devoluciones:
    - ts: Array de NumPy que contiene los valores discretos de la variable independiente.
    - sol: Array de NumPy que contiene los valores aproximados correspondientes de la variable dependiente utilizando el método de Euler.
    """
    h = (b - a) / N
    t = a
    w = yo
    ts = np.zeros(N + 1)
    sol = np.zeros(N + 1)

    for i in range(N):
        ts[i] = t
        sol[i] = w
        w = w + h * f(t, w)
        t = t + h

    ts[N] = t
    sol[N] = w

    return ts, sol
```

```python
def f(t, y):
    """
    Define la función que representa una ecuación diferencial ordinaria (EDO) de primer orden.

    Parámetros:
    - t: Variable independiente.
    - y: Variable dependiente.

    Devolución:
    - Valor de la función evaluada en los valores dados de t y y.
    """
    return y - t**2 + 1

ts, sol = met_euler(f, 0, 2, 0.5, 10)
```

```python
delta_o = np.linspace(-0.1, 0.1, 10000)
```

```python
euler_1=met_euler(f, 0, 2, 0.5 + delta_o[1], 10000)
euler_2=met_euler(f, 0, 2, 0.5 + delta_o[137], 10000)
euler_3=met_euler(f, 0, 2, 0.5 + delta_o[5000], 10000)
euler_4=met_euler(f, 0, 2, 0.5 + delta_o[368], 10000)
euler_5=met_euler(f, 0, 2, 0.5 + delta_o[-1], 10000)
euler_6=met_euler(f, 0, 2, 0.5 + delta_o[2047], 10000)
```

```python
def exacto(t):
    """
    Define la función que representa la solución exacta de una ecuación diferencial ordinaria (EDO).

    Parámetros:
    - t: Variable independiente.

    Devolución:
    - Valor de la solución exacta evaluada en el valor dado de t.
    """
    return (t + 1)**2 - 0.5 * (np.exp(t))
```
