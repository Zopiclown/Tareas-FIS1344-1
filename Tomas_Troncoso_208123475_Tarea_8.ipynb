```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def g(x):
    return (1/np.sqrt((x**2) +16))
```

Con n=10 y h=2/5

```python
x = np.linspace(0, 4, 10)
plt.fill_between(x, g(x), alpha=0.3)

h = 2/5

def poly(x):
    """
    Calcula una función polinómica a trozos en base a la variable x.

    Args:
        x (float): Valor de entrada.

    Returns:
        float: Valor de salida calculado según la función polinómica.
    """
    ind = int(x / h)
    return -g(h * ind) * (x - h * (ind + 1)) / h + g(h * (ind + 1)) * (x - h * ind) / h

plt.fill_between(x, [poly(xi) for xi in x], alpha=0.2)

xp = np.arange(9) * h
plt.scatter(xp, g(xp))
```

```python
def trapecio(f, a, b, n):
    """
    Calcula la aproximación de una integral definida utilizando la regla del trapecio.

    Args:
        f (function): La función a ser integrada.
        a (float): Límite inferior de integración.
        b (float): Límite superior de integración.
        n (int): Número de subintervalos para la aproximación. Debe ser un número entero.

    Returns:
        float: Aproximación del valor de la integral definida.
    """
    h = (b - a) / n
    fx = f(np.linspace(a + h, b - h, n - 1, endpoint=True))
    return (h / 2) * (f(a) + 2 * sum(fx) + f(b))
```

```python
trapecio(g,0,4,10)
```

```python
#error trapecio
errt = ((h**2)*4/12)*((2*((x**2) - 8)/((16+(x**2))**(5/2))))
h = np.arange(10)*h
```

```python
x = np.linspace(0, 4, 10)
plt.fill_between(x, g(x), alpha=0.3)
h = 2/5

def poly(x):
    """
    Calcula una función polinómica a trozos en base a la variable x (para fórmula compuesta de Simpson).

    Args:
        x (float): Valor de entrada.

    Returns:
        float: Valor de salida calculado según la función polinómica.
    """
    ind = int(x / (2 * h))
    Lo = (x - h * (2 * ind + 1)) * (x - h * (2 * ind + 2)) / (2 * h * h)
    L1 = -(x - h * (2 * ind)) * (x - h * (2 * ind + 2)) / (h * h)
    L2 = (x - h * (2 * ind)) * (x - h * (2 * ind + 1)) / (2 * h * h)
    return g(2 * h * ind) * Lo + g(2 * ind + 1) * L1 + g(2 * ind + 2)* L2


plt.fill_between(x, [poly(xi) for xi in x], alpha=0.2)
xp = np.arange(10) * h
plt.scatter(xp, g(xp))
plt.show()
```

```python
def simpson(f, a, b, n):
    """
    Calcula la aproximación de una integral definida utilizando la regla de Simpson compuesta.

    Args:
        f (function): La función a ser integrada.
        a (float): Límite inferior de integración.
        b (float): Límite superior de integración.
        n (int): Número de subintervalos para la aproximación. Debe ser un número entero.

    Returns:
        float: Aproximación del valor de la integral definida.
    """
    h = (b - a) / n
    fx = f(np.linspace(a + h, b - h, n - 1, endpoint=True))
    return (h / 3) * (f(a) + 2 * sum(2 * fx + 1) + 4 * sum(2 * fx - 1) + f(b))
```

```python
simpson(g,0,4,10)
```

```python
#error simpson
errs = ((h**4)*4/180)*((2*((x**2) - 8)/((16+(x**2))**(5/2))))
h = np.arange(10)*h
```

```python
plt.plot(h, errs, label = 'error metodo de simpson')
plt.plot(h, errt, label = 'error metodo del trapecio')
plt.legend()
```

Aparte de que el error en el método de Simpson se dispara en comparación al del trapecio, aproximadamente el valor real es 0.8813, por lo tanto podemos observar que la fórmula del trapecio es más precisa que la de Simpson.
